```python
# Define a class name `House`
class House: 
    
    def __init__(self):
        self.color = 'unpainted'
        self.yard = []
        
    # Set a class attribute
    # named `color`
    # assigned to 'unpainted'
    color = 'unpainted'
    
    # Set a class attribute
    # named `yard` assigned
    # to an empty list
    yard = []
    
    # Define a method named `paint`
    # include an argument named `color`
    def paint(self, color):
    
        # Update the instance's 
        # `color` attribute
        self.color = color
    
    # Define a method named `add_tree`
    def add_tree(self, tree):
        
        # Append the string 'tree'
        # to the `yard` attribute
        self.yard.append('tree')
        
    # Define a method named `add_swing_set`
    def add_swing_set(self):
        # Create a variable named `new_yard`
        # By adding the yard attribute to 
        # the list `['swing set']`
        new_yard = self.yard + ['swing set']
        
        # Assign the `yard` attribute
        # to the `new_yard` variable
        self.yard = new_yard
```

```python
#python magic methods   
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def __repr__(self):
        return f"Circle with radius: {self.radius}"

    def __add__(self, other_circle):
        return Circle(self.radius + other_circle.radius)

    def __mul__(self, factor):
        return Circle(self.radius * factor)
    
    
    
```

```python
# Polymorphism example
class Dime:
    
    value = .10
    motto = 'e pluribus unum'
    
    def __init__(self, year_minted):
        self.year_minted = year_minted
        
    
class Quarter:
    
    value = .25
    motto = 'e pluribus unum'
    
    def __init__(self, year_minted):
        self.year_minted = year_minted
```

```python
# Define a class named `Dollar`
class Dollar:
    
    # Create a class attribute
    # named `motto` assigned to the string
    # 'Liberty'
    motto = 'Liberty'
    
    # Define an __init__ method
    # that receives a `year_minted`
    # argument and sets a `year_minted`
    # instance attribute
    def __init__(self, year_minted):
        self.year_minted = year_minted
    
    # Define a method named `value`
    # that returns 4000000 if 
    # the coin was minted in 1804
    # otherwise return 1
    # https://en.wikipedia.org/wiki/Dollar_coin_(United_States)#The_1804_dollar
    # This method should behave like an attribute
    value = property(lambda self: 4000000 if self.year_minted == 1804 else 1)
```

```python
# Run for the code above to test 
from datetime import date
from random import choice

# Create coin bag
coin_bag = [Dime, Quarter, Dollar]

money = 0

# Loop over years
for year in range(1800, date.today().year):
    
    # Draw coin from bag
    coin = choice(coin_bag)
    
    # Calculate coin's value
    money += coin(year).value
    
print(f'Congratulations! You won ${money:.{2}f}!')
```
