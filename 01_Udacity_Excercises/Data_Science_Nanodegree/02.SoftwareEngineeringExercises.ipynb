```python
#python magic methods   
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def __repr__(self):
        return f"Circle with radius: {self.radius}"

    def __add__(self, other_circle):
        return Circle(self.radius + other_circle.radius)

    def __mul__(self, factor):
        return Circle(self.radius * factor)
    
    
    
```

```python
class Robot:
    
    charging_speed = .05
    
    # Create an abstract class attribute
    # named `task_battery_cost`.
    # Subclasses should set 
    # the attribute to a float
    task_battery_cost:float


    
    def __init__(self, battery=1, **kwargs):
        self.battery = battery
        self.power = False
        
    def start(self):
        self.power_on()
        self.charge()
        self.perform_task()
        self.pay_battery_cost()
        self.notify(self.power)
    
    # Define a method named `power_on`  
    def power_on(self):
    
        # Check if the robot.power is falsey
        if not self.power:
        
            # If the robot is not powered on
            # print the string 'Powering on!'
            print('Powering on!')
            
            # Set the power attribute to True
            self.power = True
            
    # Define a method named `needs_charge`
    # that checks if the battery is less than
    # the task_battery_cost
    # This method should behave like an attribute
    @property
    def needs_charge(self):
    
        # Return True if the battery attribute
        # is less than the task_battery_cost
        # otherwise return False
        return self.battery < self.task_battery_cost.float
    
    # Define a method named `charge`      
    def charge(self):
    
        # Determine if the robot needs 
        # to be charged by checking if the 
        # battery is less than the
        # cost for performing its task
        # using the `needs_charge` method
        if self.needs_charge:
        
            # If the robot needs to be charged
            # print the string 'Charging...'
            print('Charging...')
            
            # Add one charge speed to the robot's battery
            self.battery += self.charging_speed
    
    # Define an abstract method named
    # `perform_task` that receives no arguments
    def perform_task(self):
        raise NotImplementedError('Subclasses must implement this method')
        
    def pay_battery_cost(self):
        self.battery -= self.task_battery_cost
        if self.battery <= 0:
            self.power = False
           
    def notify(self, task_successful):
        if task_successful:
            print('Task completed!')
        else:
            print('Task failed!')
```

```python
# Define a subclass of `Robot` named `IndustrialRobot`.

class IndustrialRobot(Robot):
    
    # Set the charging speed to .8
    
    charging_speed = .8
    
    # Set the task_battery_cost to .5
    
    task_battery_cost = .5
    
    # Add a second parameter to the __init__ method
    # named `days_since_maintenance`
    # Make sure the parent's __init__ method is called
    
    def __init__(self, battery=1.0, days_since_maintenance=0, **kwargs):
    
        # Set an instance attribute named `days_since_maintenance`
        # Call the parent class's __init__ method
        
        super().__init__(battery=battery, **kwargs) 
        self.days_since_maintenance = days_since_maintenance
        

        
    def perform_task(self):
        print('beep boop beep!')
    
    # Overwrite the notify method
    # to check if 60 days has passed since
    # the robot received maintenance and
    # print the string 'Requested maintenance!'
    # Ensure the parent class's notify method is
    # called
    
    def notify(self, task_successful):
        
        # Check if it's been 60 days since maintenance
        
        if self.days_since_maintenance >= 60:
        
            # Print the string "Requested maintenance!!"
            
            print("Requested maintenance!!")

        
        # Call the parent class's `notify` method
     
        super().notify(task_successful)
```
