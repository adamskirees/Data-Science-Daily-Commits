```python
# " Self" refers to the instance of the class itself.
# It allows you to access attributes and methods of the class in Python.

# Define a class named `A`
class A:


    # Define a method named `return_self`
    def return_self(self):
    
        # Return self
        return self
    
# Define an instance of A
# assigned to the variable `a`
a = A()

# Call the `return_self` method
# and assign the output
# to a variable named `a_self`
a_self = a.return_self()

print(a == a_self)
print(a == A)
print(a == A())
print(a)
print(a_self)
print(a_self.return_self())
```
